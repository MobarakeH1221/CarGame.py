````
import cv2
import numpy as np
import requests
from PIL import Image
from io import BytesIO
import matplotlib.pyplot as plt

# URL of the image
image_url = "https://images.pexels.com/photos/376464/pexels-photo-376464.jpeg?auto=compress&cs=tinysrgb&dpr=1&w=500"

# Fetch the image from the URL
response = requests.get(image_url)
image_bytes = BytesIO(response.content)

# Read the image using PIL (Python Imaging Library)
image_pil = Image.open(image_bytes)

# Convert PIL image to OpenCV format
image_cv = cv2.cvtColor(np.array(image_pil), cv2.COLOR_RGB2BGR)

# Edge Detection using Canny edge detector
edges = cv2.Canny(image_cv, 100, 200)

# Splitting RGB channels
red_channel = image_cv[:, :, 0]
green_channel = image_cv[:, :, 1]
blue_channel = image_cv[:, :, 2]

# Displaying the original image
plt.subplot(2, 3, 1)
plt.imshow(cv2.cvtColor(image_cv, cv2.COLOR_BGR2RGB))
plt.title('Original Image')
plt.axis('off')

# Displaying the edge-detected image
plt.subplot(2, 3, 2)
plt.imshow(edges, cmap='gray')
plt.title('Edge Detected')
plt.axis('off')

# Displaying the red channel
plt.subplot(2, 3, 4)
plt.imshow(red_channel, cmap='gray')
plt.title('Red Channel')
plt.axis('off')

# Displaying the green channel
plt.subplot(2, 3, 5)
plt.imshow(green_channel, cmap='gray')
plt.title('Green Channel')
plt.axis('off')

# Displaying the blue channel
plt.subplot(2, 3, 6)
plt.imshow(blue_channel, cmap='gray')
plt.title('Blue Channel')
plt.axis('off')

plt.tight_layout()
plt.show()
````

````
import cv2
import numpy as np
import requests
from PIL import Image
from io import BytesIO
import matplotlib.pyplot as plt
from skimage.feature import hog
from skimage import exposure

# URL of the image
image_url = "https://images.pexels.com/photos/376464/pexels-photo-376464.jpeg?auto=compress&cs=tinysrgb&dpr=1&w=500"

# Fetch the image from the URL
response = requests.get(image_url)
image_bytes = BytesIO(response.content)

# Read the image using PIL (Python Imaging Library)
image_pil = Image.open(image_bytes)

# Convert PIL image to grayscale
image_gray = image_pil.convert('L')

# Convert PIL image to OpenCV format
image_cv = cv2.cvtColor(np.array(image_pil), cv2.COLOR_RGB2BGR)

# Compute HOG features
fd, hog_image = hog(image_gray, orientations=9, pixels_per_cell=(8, 8), cells_per_block=(2, 2), visualize=True)

# Rescale histogram for better visual representation
hog_image_rescaled = exposure.rescale_intensity(hog_image, in_range=(0, 10))

# Displaying the original image
plt.subplot(1, 2, 1)
plt.imshow(image_cv)
plt.title('Original Image')
plt.axis('off')

# Displaying the HOG image
plt.subplot(1, 2, 2)
plt.imshow(hog_image_rescaled, cmap='gray')
plt.title('HOG Features')
plt.axis('off')

plt.tight_layout()
plt.show()
````

````
import cv2
import numpy as np
import requests
from PIL import Image
from io import BytesIO
import matplotlib.pyplot as plt

# URL of the image
image_url = "https://images.pexels.com/photos/376464/pexels-photo-376464.jpeg?auto=compress&cs=tinysrgb&dpr=1&w=500"

# Fetch the image from the URL
response = requests.get(image_url)
image_bytes = BytesIO(response.content)

# Read the image using PIL (Python Imaging Library)
image_pil = Image.open(image_bytes)

# Convert PIL image to OpenCV format
image_cv = cv2.cvtColor(np.array(image_pil), cv2.COLOR_RGB2BGR)

# Convert image to HSV color space
hsv_image = cv2.cvtColor(image_cv, cv2.COLOR_BGR2HSV)

# Define color filter parameters (ranges)
lower_blue = np.array([90, 50, 50])  # Lower bound for blue color in HSV
upper_blue = np.array([130, 255, 255])  # Upper bound for blue color in HSV

# Create mask for blue color
mask_blue = cv2.inRange(hsv_image, lower_blue, upper_blue)

# Apply mask to original image to extract blue color
blue_filtered_image = cv2.bitwise_and(image_cv, image_cv, mask=mask_blue)

# Displaying the original image
plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(image_cv, cv2.COLOR_BGR2RGB))
plt.title('Original Image')
plt.axis('off')

# Displaying the blue-filtered image
plt.subplot(1, 2, 2)
plt.imshow(cv2.cvtColor(blue_filtered_image, cv2.COLOR_BGR2RGB))
plt.title('Blue Filtered Image')
plt.axis('off')

plt.tight_layout()
plt.show()
````

````
import cv2
import numpy as np
import requests
from PIL import Image
from io import BytesIO
import matplotlib.pyplot as plt

# URL of the image
image_url = "https://images.pexels.com/photos/376464/pexels-photo-376464.jpeg?auto=compress&cs=tinysrgb&dpr=1&w=500"

# Fetch the image from the URL
response = requests.get(image_url)
image_bytes = BytesIO(response.content)

# Read the image using PIL (Python Imaging Library)
image_pil = Image.open(image_bytes)

# Convert PIL image to OpenCV format
image_cv = cv2.cvtColor(np.array(image_pil), cv2.COLOR_RGB2BGR)

# Define the region of interest (ROI) for cropping
x, y, w, h = 100, 100, 300, 200  # Coordinates and dimensions of the ROI

# Crop the image
cropped_image = image_cv[y:y+h, x:x+w]

# Displaying the original image
plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(image_cv, cv2.COLOR_BGR2RGB))
plt.title('Original Image')
plt.axis('off')

# Displaying the cropped image
plt.subplot(1, 2, 2)
plt.imshow(cv2.cvtColor(cropped_image, cv2.COLOR_BGR2RGB))
plt.title('Cropped Image')
plt.axis('off')

plt.tight_layout()
plt.show()
````
